```python
import cv2
import matplotlib.pyplot as plt
```

```python
def vid_cap(nuum = 0):
    cap = cv2.VideoCapture(nuum)
    if cap.isOpened():
        h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        cap.release()
        return nuum,h,w
    else:
        return nuum,None,None
```

```python
for i in range(10):
    nuum,h,w = vid_cap(i)
    print(f'num : {nuum} \n  ---- \n         hight : {h}\n ----  \n         width : {w}')
```

```python
def start_camera(s = 0):
    i = 0
    if s==0:
        cap = cv2.VideoCapture(0)
        while cap.isOpened():
            ans,frame = cap.read()
            if ans:
#                 results = model.predict(frame)
                cv2.imshow("Camera", frame)
            key  = cv2.waitKey(1)
            if key == ord("q"):
                break
            elif key == ord("s"):
                cv2.imwrite(f'data/img{i}.png',frame)
                i += 1
                print(i)
                
        cap.release()
        cv2.destroyAllWindows()
    else:
        print(None)
```

```python
start_camera(0)
```
